```python
import requests
from bs4 import BeautifulSoup
```

```python
#取得将html经过处理了的soup.
res = requests.get('http://tech.sina.com.cn/mobile/n/n/2018-01-16/doc-ifyqqieu6764690.shtml')
res.encoding='utf-8'
soup= BeautifulSoup(res.text,'html.parser')
#print(soup)
```

```python
#取得新闻标题
#id 的 select('#....')
soup.select('#artibodyTitle')[0].text
```

```python
#取得新闻发布时间
timesource=soup.select('#pub_date')[0].text
timesource
```

```python
#日期按格式转化为datetime元组
from datetime import datetime
dt=datetime.strptime(timesource,'%Y年%m月%d日%H:%M')
dt
```

```python
#转化为字符串
dt.strftime('%Y-%m-%d')
```

```python
#取得新闻来源
#指定类或id下的行级标签
soup.select('#media_name  a')[0].text
```

```python
#取得新闻内容
#查询所有id=artibody 下的 p标签
soup.select('#artibody p')
```

```python
#将段落组成文章
article= []
for p in soup.select('#artibody p'):
    article.append(p.text.strip())
print(article)
```

```python
#将段落隔开
print('\n\n'.join(article))
```

```python
#截取,匹配字符串中的子串
#比如需要取得文章编号fyqrewi9270442
newsurl='http://tech.sina.com.cn/mobile/n/n/2018-01-17/doc-ifyqrewi9270442.shtml'
#使用字符串bif进行切割
newsid=newsurl.split('/')[-1].lstrip('doc-i').rstrip('.shtml')
newsid
```

```python
#使用re套件
import re
m=re.search('doc-i(.+).shtml',newsurl)
newsid=m.group(1)
newsid
```
