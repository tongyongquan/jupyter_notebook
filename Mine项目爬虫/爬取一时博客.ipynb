```python
import requests
from bs4 import BeautifulSoup
```

```python
#取得页面html
url='http://hellogod.cn/'
res = requests.get(url)
soup= BeautifulSoup(res.text,'html.parser')
print(soup)
```

```python
import re
soup.select('link')
```

```python
#获得所有链接
[link.get('href') for link in soup.select('link')]
```

```python
css=[]
hrefs=[link.get('href') for link in soup.select('link')]
for href in hrefs:
    if 'css' in href:
        print('GET '+href.split('/')[-1])
        if'http' not in href:
            href='http://hellogod.cn'+href
        res_css=requests.get(href)
        css.append(res_css)
print(css)
```

```python
css[1]=requests.get('http://fonts.proxy.ustclug.org/css?family=Lato:300,300italic,400,400italic,700,700italic&subset=latin,latin-ext')
print(css)
```

```python
#得到所有style
style=''
for sty in css :
    style+=sty.text
    
style='<style>'+style+'</style>'

print(style)
```
